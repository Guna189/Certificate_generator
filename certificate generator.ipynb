```python
import cv2 as cv
import openpyxl
from PIL import Image
import camelcase as cc
```

```python
certificate_template="TEMPLATE.png"
certified_names="EXCELSHHET.xlsx"
generated_certificates="generated/"

font_size=6
font_color=(0,0,0)

#no.of certificates
n=99

#Certificate name Co-ordinates
x_corr=150
y_corr=0

obj = openpyxl.load_workbook(certified_names)
sheet = obj.active

success,failed=[],[]
for i in range(1,n+1):
    try:
        get_name = sheet.cell(row = i ,column = 1)
        certi_name = get_name.value
        certi_name = certi_name.title()
        print(certi_name,end="\t")

        img = cv.imread(certificate_template)

        #certificate enlarger
        p = 2
        w = img.shape[1] * p
        h = img.shape[0] * p
        img = cv.resize(img, (w, h))


        #img font-style
        font = cv.FONT_HERSHEY_SCRIPT_COMPLEX

        text_size = cv.getTextSize(certi_name, font, font_size,0)[0]

        text_x = (img.shape[1] - text_size[0]) / 2 + x_corr 
        text_y = (img.shape[0] + text_size[1]) / 2 - y_corr
        text_x = int(text_x)
        text_y = int(text_y)
        #print(text_x,text_y)
        cv.putText(img, certi_name,(text_x ,text_y ), font,font_size,font_color, 10)

        #participant img import
        get_id = sheet.cell(row = i ,column = 2)
        certi_roll = get_id.value

        #img_path=certi_roll+".png"
        img_path = "/PATH.PNG"
        img2=cv.imread(img_path)
        #participant img resize
        w = 330
        h = 440
        dim = (w,h)
        img2 = cv.resize(img2,dim)


        #participant img co-ordinates
        yc=1000
        xc=3350
        rows,cols,channels = img2.shape
        roi = img[yc:rows+yc, xc:cols+xc]


        img2gray = cv.cvtColor(img2, cv2.COLOR_BGR2GRAY)
        ret, mask = cv.threshold(img2gray, -255, 255, cv2.THRESH_BINARY)
        mask_inv = cv.bitwise_not(mask)

        img1_bg = cv.bitwise_and(roi, roi, mask=mask_inv)

        img2_fg = cv.bitwise_and(img2, img2, mask=mask)

        dst = cv.add(img1_bg, img2_fg)


        img[yc:rows+yc, xc:cols+xc] = dst


        #final certificate shrinker
        p = 0.5
        w = int(img.shape[1] * p)
        h = int(img.shape[0] * p)
        img = cv.resize(img, (w, h))


        #img renaming
        certi_path = generated_certificates+ '/' + str(certi_name) + '.png'
        cv.imwrite(certi_path,img)

        #img to pdf
        img = Image.open(certi_path)
        img = img.convert('RGB')
        img.save(generated_certificates+"pdfs/"+ str(certi_name) +'.pdf')
        print(i," completed")
        success.append(i)
    except:
        failed.append(i)
        
        
        
print("Succes: ",len(success),success)
print("Failed: ",len(failed),failed)
```
